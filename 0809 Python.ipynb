# http://www.hani.co.kr/rss/sports/ 에서 title 과 link 태그의 내용을 읽어서 DataFrame 생성

```python
import xml.etree.ElementTree as et # XML 파싱을 위한 import
import urllib.request # 웹에서 문자열을 다운로드 받기 위한 import
```

```python
# 데이터 다운로드
# 파싱과 관계 없이 텍스트를 가지고 옴
url = 'http://www.hani.co.kr/rss/sports/'
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
#print(response.read())
```

```python
# parsing 을 위해서는 root 를 찾아야 함
tree = et.parse(response)
xroot = tree.getroot()
print(xroot)
```

```python
# 찾고자 하는 태그의 내용을 가져오기
channel = xroot.findall('channel')
#print(channel)
```

```python
# channel 아래의 item 태그의 내용 가져오기
# channel 은 데이터 1개만 존재하지만 list 형식이므로 인덱스 0 을 사용
items = channel[0].findall('item')
print(items)
```

```python
# item 태그의 데이터들 확인 및 dict 데이터 생성
rows = []
for item in items:
    s_title = item.find('title').text
    s_link = item.find('link').text
    #print(s_title, s_link)
    rows.append({'title' : s_title, 'link' : s_link})

# DataFrame 으로 변환
DF = pd.DataFrame(rows, columns = ['title', 'link'])
print(DF)
    
```
